# Web Scraping in Python

**Problem Statement:** We are going scrape Flipkart website to extract the Price, Name, and Rating of Laptops. The URL for this page is https://www.flipkart.com/laptops/~buyback-guarantee-on-laptops-/pr?sid=6bo%2Cb5g&uniqBStoreParam1=val1&wid=11.productCard.PMU_V2.

### Step1: import all the necessary libraries:

```python
# importing the libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Collecting and Parsing a Webpage
In the next step, we will make a GET request to the url and will create a object(soup) with the help of BeautifulSoup.

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\chromedriver.exe") #used to access Chrome

products=[] #List to store name of the product
prices=[] #List to store price of the product
ratings=[] #List to store rating of the product

driver.get("https://www.flipkart.com/laptops/~buyback-guarantee-on-laptops-/pr?sid=6bo%2Cb5g&uniq")
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
print(soup.prettify()) # print the parsed data of html
```

```python
#print title of webpage
print(soup.title)
```

```python
for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}):
    name=a.find('div', attrs={'class':'_3wU53n'})
    price=a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
    rating=a.find('div', attrs={'class':'hGSR34'})
    products.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text) 
```

```python
df = pd.DataFrame({'Product Name':products,'Price':prices,'Rating':ratings}) 
df.to_csv('products.csv', index=False, encoding='utf-8')

df
```
